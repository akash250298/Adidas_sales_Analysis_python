```python
# group sales region wise
region_sales = df.groupby('Region')['Total Sales'].sum().sort_values(ascending=False)
region_sales.head(10)
```

```python
# Convert 'Invoice Date' to datetime if not already
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])

monthly_sales = df.groupby(df['Invoice Date'].dt.to_period('M'))['Total Sales'].sum()
monthly_sales.head()
```

```python
# Remove possible leading/trailing whitespaces from columns
df.columns = df.columns.str.strip()

# Convert 'Total Sales' to numeric, force errors to NaN
df['Total Sales'] = pd.to_numeric(df['Total Sales'], errors='coerce')
```

```python
# Region-wise total sales
region_sales = df.groupby('Region')['Total Sales'].sum().sort_values(ascending=False)
print(region_sales)
```

### Discover `Price per Unit` column

```python
df['Price per Unit'].unique()
```

```python
# remove $ 

def removeDollarSign(price):
    if isinstance(price, str):
        return price.replace('$', '')
    return price  # Return as is if not a string (e.g., NaN)

df['Price per Unit'] = df['Price per Unit'].apply(removeDollarSign)
df['Price per Unit']
```

```python
# convert data type to numeric

df['Price per Unit'] = pd.to_numeric(df['Price per Unit'], errors = 'coerce')
```

```python
# check nan value after assign error equal to "coerce"

df['Price per Unit'].isna().sum()
```

```python
sns.boxplot(y = 'Price per Unit', data = df)
plt.show()
```

```python
# Drop some columns that we don't need

df.drop(columns=['Correct total', 'profit_percentage'], inplace=True, errors='ignore')
df.sample(3)
```

```python
# Show Total Sales and Operating Profit

region_sales_percentage = ((region_sales / region_sales.sum()) * 100).round()
region_sales_df = pd.DataFrame({
	'Total Sales': region_sales,
	'Sales Percentage': region_sales_percentage
})
region_sales_df
```

```python
# What is total sales, Operating Profit in each Region ?

# Aggregate Operating Profit by region (cleaning $ and ,)
df['Operating Profit Clean'] = (
	df['Operating Profit']
	.str.replace('$', '')
	.str.replace(',', '')
	.astype(float)
)
region_profit = df.groupby('Region')['Operating Profit Clean'].sum().sort_values(ascending=False)

plt.figure(figsize=(15, 6))

# Plot Total Sales
plt.subplot(1, 2, 1)
sns.lineplot(x=region_sales_df.index, y=region_sales_df["Total Sales"], marker="o")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

# Plot Operating Profit
plt.subplot(1, 2, 2)
sns.lineplot(x=region_profit.index, y=region_profit.values, marker="o")
plt.title("Operating Profit by Region")
plt.xlabel("Region")
plt.ylabel("Operating Profit")

plt.tight_layout()
plt.show()
```

- `West` and `Northeast` emerge as the top Region, capture 30% and 21% of sales  respectively.

```python
# Show Total Sales and Operating Profit by Sales Method
sales_method_df = df.groupby('Sales Method').agg({
	'Total Sales': 'sum',
	'Operating Profit': lambda x: x.str.replace('$', '').str.replace(',', '').astype(float).sum()
}).reset_index()

sales_method_df['sales percentage'] = ((sales_method_df['Total Sales'] / sales_method_df['Total Sales'].sum())*100).round(2)
sales_method_df
```

```python
df['Year'] = df['Invoice Date'].dt.year
df['Month'] = df['Invoice Date'].dt.month

sales_by_month = df.groupby(['Year', 'Month'])['Total Sales'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x="Month", y="Total Sales", hue="Year", data=sales_by_month, marker='o')

plt.show()
```

- In `2021`, sales were much higher than in `2020`, and the main reason for this was Covid-19.
